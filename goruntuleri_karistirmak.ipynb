```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(filename='img1.jpg')
img2 = cv2.imread(filename='img2.jpg')
```

```python
plt.figure()
plt.imshow(X=img1);
```

```python
plt.figure()
plt.imshow(X=img2);
```

### resimlerinin renklerinin böyle görünmesinin sebebi, opencv bgr renk skalasını kullanıyor. Fakat üzerinde çalıştığımız renkler başta olmak üzere neredeyse her görüntü dosyası rgb renk skalasını kullanıyor. Bu sorunu düzeltmek için dönüştürme işlemi yapabiliriz.

```python
img1 = cv2.cvtColor(src=img1, code=cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(X=img1);
```

```python
img2 = cv2.cvtColor(src=img2, code=cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(X=img2);
```

### Birden fazla görüntüyü birbiri ile karıştırmak istediğimiz zaman bu görüntülerin boyutlarının birbirlerine eşit olması gerekir. Yoksa hata alırız. Bunun için eşitleme işlemi yapabiliriz.

```python
img1.shape, img2.shape
```

```python
img1 = cv2.resize(src=img1, dsize=(600, 600))
img2 = cv2.resize(src=img2, dsize=(600, 600))
img1.shape, img2.shape
```

```python
plt.figure()
plt.imshow(X=img1);
```

```python
plt.figure()
plt.imshow(X=img2);
```

```python
blended = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
plt.figure()
plt.imshow(X=blended);
```

```python
blended1 = cv2.addWeighted(src1=img1, alpha=0.8, src2=img2, beta=0.2, gamma=0)
plt.figure()
plt.imshow(X=blended1);
```

```python
blended2 = cv2.addWeighted(src1=img1, alpha=0.2, src2=img2, beta=0.8, gamma=0)
plt.figure()
plt.imshow(X=blended2);
```
