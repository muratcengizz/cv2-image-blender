```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(filename='img1.jpg')
img2 = cv2.imread(filename='img2.jpg')
```

```python
plt.figure()
plt.imshow(X=img1);
```

```python
plt.figure()
plt.imshow(X=img2);
```

### resimlerinin renklerinin byle grnmesinin sebebi, opencv bgr renk skalasn kullanyor. Fakat zerinde altmz renkler bata olmak zere neredeyse her grnt dosyas rgb renk skalasn kullanyor. Bu sorunu dzeltmek iin dntrme ilemi yapabiliriz.

```python
img1 = cv2.cvtColor(src=img1, code=cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(X=img1);
```

```python
img2 = cv2.cvtColor(src=img2, code=cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(X=img2);
```

### Birden fazla grnty birbiri ile kartrmak istediimiz zaman bu grntlerin boyutlarnn birbirlerine eit olmas gerekir. Yoksa hata alrz. Bunun iin eitleme ilemi yapabiliriz.

```python
img1.shape, img2.shape
```

```python
img1 = cv2.resize(src=img1, dsize=(600, 600))
img2 = cv2.resize(src=img2, dsize=(600, 600))
img1.shape, img2.shape
```

```python
plt.figure()
plt.imshow(X=img1);
```

```python
plt.figure()
plt.imshow(X=img2);
```

```python
blended = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
plt.figure()
plt.imshow(X=blended);
```

```python
blended1 = cv2.addWeighted(src1=img1, alpha=0.8, src2=img2, beta=0.2, gamma=0)
plt.figure()
plt.imshow(X=blended1);
```

```python
blended2 = cv2.addWeighted(src1=img1, alpha=0.2, src2=img2, beta=0.8, gamma=0)
plt.figure()
plt.imshow(X=blended2);
```
